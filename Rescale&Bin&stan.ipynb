```python
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
irisdata = datasets.load_iris()
irisdata
```

```python
x = irisdata.data
y = irisdata.target
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
mm = MinMaxScaler()
```

```python
x1 = mm.fit_transform(x_train)
x2 = mm.transform(x_test)
```

```python
x1
```

```python
# Binarize Data
```

```python
from sklearn.preprocessing import Binarizer
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
colnames = ["preg","plas","pres","skin","ins","mass","depi","age","class"]
```

```python
data = pd.read_csv(url,names=colnames)
data
```

```python
array = data.values
array
```

```python
x = array[:,0:8]
y = array[:,8]
```

```python
x
```

```python
y
```

```python
binarize = Binarizer(threshold=0.0).fit(x)
```

```python
ans = binarize.transform(x)
```

```python
ans[0:10,:]
```

```python
# Standardize Data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(x)
```

```python
rescale = scaler.transform(x)
rescale[0:10,:]
```
